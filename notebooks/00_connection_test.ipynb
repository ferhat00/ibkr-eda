# IBKR Connection Test

Verify that IB Gateway is running and connected.

**Prerequisites:**
1. Start IB Gateway (`C:\Jts\ibgateway\1044\ibgateway.exe`)
2. Log in with your IBKR credentials
3. Ensure API connections are enabled (Configure → Settings → API → Enable ActiveX and Socket Clients)
4. Default port: 4002 (paper trading) or 4001 (live)

```python
from ibkr_eda import IBKR

ib = IBKR()  # auto-connects to IB Gateway
```

## Connection Status

```python
ib.status()
```

## Accounts

```python
ib.accounts.list_accounts()
```

## Account Summary

```python
ib.accounts.get_summary()
```

## Positions

```python
positions = ib.positions.get()
positions
```

## Recent Trades

```python
ib.executions.get()
```

## Contract Search (example: AAPL)

```python
ib.contract_search.search("AAPL")
```

## Historical Data (example: AAPL, conid=265598)

```python
history = ib.history.get(conid=265598, period="1m", bar="1d")
history
```

## Asset Allocation

```python
allocation = ib.accounts.get_allocation()
for category, df in allocation.items():
    print(f"\n--- {category} ---")
    display(df)
```

## Disconnect

```python
ib.disconnect()
```
