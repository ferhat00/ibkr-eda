# IBKR Connection Test

Verify that IB Gateway is running and connected.

**Prerequisites:**
1. Start IB Gateway (`C:\Jts\ibgateway\1044\ibgateway.exe`)
2. Log in with your IBKR credentials
3. Ensure API connections are enabled (Configure → Settings → API → Enable ActiveX and Socket Clients)
4. Default port: 4002 (paper trading) or 4001 (live)

```python
from ibkr_eda import IBKR

ib = await IBKR.create_async()  # async connect for Jupyter / Python 3.14+
```

## Connection Status

```python
ib.status()
```

## Accounts

```python
ib.accounts.list_accounts()
```

## Account Summary

## Positions

```python
positions = ib.positions.get(account_id="U14593335")
positions
```

## Recent Trades

```python
await ib.executions.get_async(account_id="U14593335", start_date="2025-01-01", end_date="2026-03-02")
```

## Contract Search (example: AAPL)

```python
await ib.contract_search.search_async("AAPL")
```

## Historical Data (example: AAPL, conid=265598)

```python
history = await ib.history.get_async(conid=265598, period="1m", bar="1d")
history
```

## Asset Allocation

```python
allocation = ib.accounts.get_allocation(account_id="U14593335")
for category, df in allocation.items():
    print(f"\n--- {category} ---")
    display(df)
```

## Disconnect

```python
ib.disconnect()
```
