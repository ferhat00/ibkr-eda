# IBKR Connection Test

Verify that the Client Portal Gateway is running and authenticated.

**Prerequisites:**
1. Start the IBKR Client Portal Gateway
2. Log in via browser at https://localhost:5000
3. Copy `.env.example` to `.env` and set your account ID (optional)

```python
from ibkr_eda import IBKR

ib = IBKR()
ib.keepalive()
```

## Authentication Status

```python
ib.status()
```

## Accounts

```python
ib.accounts.list_accounts()
```

## Account Summary

```python
ib.accounts.get_summary()
```

## Positions

```python
positions = ib.positions.get()
positions
```

## Recent Trades

```python
ib.executions.get()
```

## Contract Search (example: AAPL)

```python
ib.contract_search.search("AAPL")
```

## Historical Data (example: AAPL, conid=265598)

```python
history = ib.history.get(conid=265598, period="1m", bar="1d")
history
```

## Asset Allocation

```python
allocation = ib.accounts.get_allocation()
for category, df in allocation.items():
    print(f"\n--- {category} ---")
    display(df)
```

## Cleanup

```python
ib.stop_keepalive()
```
